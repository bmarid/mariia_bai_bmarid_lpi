```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
link = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df = pd.read_csv(link)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df.date, format='%Y-%m-%d')
```

```python
df.describe()
```

```python
df.columns
```

# DATA CLEANING
### Separate columns on parts, drop empty

```python
part_1_general = ['iso_code', 'continent', 'location', 'date', 'total_cases', 'new_cases']
       
part_2_cases = ['new_cases_smoothed', 'total_deaths', 'new_deaths', 'new_deaths_smoothed', 'reproduction_rate']

part_3_per_million = ['total_cases_per_million', 'new_cases_per_million', 'new_cases_smoothed_per_million',
                   'total_deaths_per_million', 'new_deaths_per_million', 'new_deaths_smoothed_per_million']

part_4_patients = ['icu_patients', 'icu_patients_per_million', 'hosp_patients','hosp_patients_per_million', 
        'weekly_icu_admissions','weekly_icu_admissions_per_million', 'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million']
          
part_5_tests = ['total_tests', 'new_tests', 'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand','positive_rate', 'tests_per_case', 'tests_units']

part_6_vaccine = ['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters',
       'new_vaccinations', 'new_vaccinations_smoothed', 'total_vaccinations_per_hundred', 
       'people_vaccinated_per_hundred',  'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred',
       'new_vaccinations_smoothed_per_million','new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred']

part_7_other = ['stringency_index', 'population', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty','cardiovasc_death_rate', 'diabetes_prevalence', 
       'female_smokers', 'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand','life_expectancy',
        'human_development_index']

part_8_ex_mr = ['excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
       'excess_mortality', 'excess_mortality_cumulative_per_million']
```

```python
df[part_8_ex_mr].tail()
```

```python
for i in part_8_ex_mr:
    print(i,df[i].values)
```

```python
# If no results, we can drop columns
to_drop = part_4_patients + part_5_tests + part_6_vaccine + part_8_ex_mr
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

### Preparing data

```python
df['year'] = df['date'].dt.strftime('%Y')
```

```python
df['month_year'] = df['date'].dt.strftime('%m-%Y')
```

```python
df['quarter'] = pd.PeriodIndex(df.date, freq='Q')
```

```python
quarter_values = df['quarter'].value_counts()
quarter_values
```

```python
# Bars are good for segmentation of information
# We can see the amout per quarter, how amount is going
plt.figure(figsize=(20,5))
sns.countplot(x='quarter', data = df)
plt.title('Countplot')
```

```python
# Get values (list) of continent column
cont = df['continent'].value_counts().to_frame()
label_cont_list = cont.index.to_list()
label_cont_list
```

```python
# Pie chart - to show parts as a percentage of each other
# this represents % of days in continents with cases
labels = label_cont_list
sizes = df['continent'].value_counts()

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Pie chart by continents')
```

```python
# Same, more specific
sns.set(style='darkgrid')
df['continent'].value_counts().plot(kind="bar")
```

```python
# To show the correlation between new cases and date
# New cases density
figure(figsize=(20, 3))
plt.scatter(df['date'], df['new_cases'], color = 'blue')
plt.title('Scatter plot for cases analysys')
plt.xlabel('date')
plt.ylabel('new_cases')
```

```python
# Orange line higher than blue
# Males smokes more than females
# data as value
values = df[['female_smokers','male_smokers']]
figure(figsize=(20, 3))
plt.plot(values)
plt.legend(['female_smokers','male_smokers'])
plt.title('Example of Line chart')
```

```python
# Same in barplot
sns.barplot(df[['male_smokers','female_smokers']])
plt.title('Barplot')
```

```python
# Amount of new cases per year
sns.boxplot(x = 'new_cases', y = 'quarter', data = df)
plt.title('Boxplot')
```

```python
# Displot shows if there is a connection between 
# Looks like scatter plot, but because of rounds we can see full-connected and finished figure
sns.displot(data=df, x="diabetes_prevalence", y="human_development_index", kind="kde")
```
